```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter
from IPython.display import Image, display

fps = 20
duration = 12
dt = 1 / fps
frames = int(duration * fps)

A0 = 1.0
growth_k = 0.08
beat_base = 0.55
beat_amp = 0.45
bpm = 72

omega = 2 * np.pi * (bpm / 60)
theta = np.linspace(0, 2 * np.pi, 800)

def heart_xy(scale=1.0):
    x = 16 * np.sin(theta) ** 3
    y = (
        13 * np.cos(theta)
        - 5 * np.cos(2 * theta)
        - 2 * np.cos(3 * theta)
        - np.cos(4 * theta)
    )
    return scale * x, scale * y

def heartbeat(t):
    s = np.sin(omega * t)
    pulse = np.maximum(0, s) ** 6
    return beat_base + beat_amp * pulse

fig, ax = plt.subplots(figsize=(7, 8))
ax.set_aspect("equal")
ax.axis("off")
ax.set_xlim(-32, 32)
ax.set_ylim(-42, 34)

heart_line, = ax.plot([], [], lw=2.5, color="red")

# data readout
text_box = ax.text(
    0.05, 0.93, "",
    transform=ax.transAxes,
    va="top",
    family="monospace",
    fontsize=10
)

# poetic title
title_text = ax.text(
    0.5, 0.985,
    "Can You Differentiate My Love?",
    transform=ax.transAxes,
    ha="center",
    va="top",
    fontsize=15,
    fontweight="bold"
)

# nerdy-poetic body text
poem_text = ax.text(
    0.5, 0.14,
    "My love, if quantifiable, would not be static or transient.\n"
    "Love is not a point in time, but a function of growth in time.\n"
    "Not a single heartbeat, but the integration of many.\n"
    "With every heartbeat it continues to grow.\n"
    r"$A(t) = A_0 + k\int_{0}^{t} h(\tau)\,d\tau$",

    transform=ax.transAxes,
    ha="center",
    va="center",
    fontsize=11,
    style="italic"
)

def update(frame):
    t = frame * dt
    cumulative = 0.0
    for i in range(frame + 1):
        cumulative += heartbeat(i * dt) * dt

    A = A0 + growth_k * cumulative
    scale = np.sqrt(A / A0)

    x, y = heart_xy(scale)
    heart_line.set_data(x, y)

    text_box.set_text(
        f"time:            {t:5.2f}s\n"
        f"heartbeat:       {heartbeat(t):5.3f}\n"
        f"integrated love: {cumulative:5.3f}\n"
        f"heart area:      {A:5.3f}"
    )

    return heart_line, text_box, title_text, poem_text

anim = FuncAnimation(
    fig,
    update,
    frames=frames,
    interval=1000 / fps,
    blit=False
)

gif_path = "love_heart_poem.gif"
anim.save(gif_path, writer=PillowWriter(fps=fps))
plt.close(fig)

display(Image(filename=gif_path))
```
